## Step 2: Fit model

### 2.1 Define algorithm

```python
from time import time

assert X_train.shape[0] == y_train.shape[0]

theta = np.zeros(X_train.shape[1]) # initial weight (theta)
lr = 0.1
epochs = 50

def h_theta(X, theta):
    return X @ theta

def mse(yhat, y):
    m = yhat.shape[0] # m: number of samples
    return ((yhat - y) ** 2).sum() / m
    # return ((yhat - y) ** 2).sum() / yhat.shape[0]


def r2(yhat, y):
    ssr = ((y - yhat) ** 2).sum() # ssr: residual sum of squares
    ybar = y.mean()
    sst = ((y - ybar) ** 2).sum() # sst: total sum of squares
    r2_score = 1 - (ssr / sst)
    return r2_score

def gradient(X, error):
    m = X.shape[0] # m: number of samples
    return (X.T @ error) / m
    # return (X.T @ error) / X.shape[0]


# ------------------------- fitting area ------------------------
start_time = time()

for i in range(epochs):

    print(f'epoch: {i}')

    # Step 1: Predict
    # 1. yhat = X @ w
    # prediction
    # yhat (m, ) = (m, n) @ (n, )

    # Add for I want to know
    print(f'X_train = {X_train}')
    print(f'X_train.shape = {X_train.shape}')
    print(f'theta = {theta}')
    print(f'theta.shape = {theta.shape}')

    yhat = h_theta(X_train, theta)

    # Add for I want to know
    print(f'yhat    = {yhat}')
    print(f'yhat.shape = {yhat.shape}')
    print(f'y_train = {y_train}')
    print(f'y_train.shape = {y_train.shape}')

    # Step 2: Error
    # error for use to calculate gradients
    # error (m, ) = (m, ) - (m, )
    error = yhat - y_train
    print(f'error = {error}')
    print(f'error.shape = {error.shape}')

    # Step 3: Gradient
    # grad (n, ) = (n, m) @ (m, )
    # grad for each feature j
    grad = gradient(X_train, error)
    print(f'grad = {grad}')
    print(f'grad.shape = {grad.shape}')

    # Step 4: Update w
    # w = w - lr * grad
    # w (n, ) = (n, ) - scalar * (n, )
    theta = theta - lr * grad
    print(f'update theta: \n{theta}')
    print(f'update theta.shape = {theta.shape}')
    
    print('----------------------------')

time_taken = time() - start_time
```

## 2.2 Compute accuracy/loss

```python
# Now got lovely w, it's time to check our accuracy

# 1. Make prediction
yhat = h_theta(X_test, theta)

# 2.1 Calculate mean squared errors
mse = mse(yhat, y_test)

# 2.2 Calculate r2 score
r2 = r2(yhat, y_test)

# Print the mse and r2
print('MSE: ', mse)shape
print('r2: ', r2)
print("Time used: ", time_taken)
```
